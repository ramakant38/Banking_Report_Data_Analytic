```python
bins = [0, 100000, 300000, float('inf')]
labels = ['low','med','high']

df1['income band'] = pd.cut(df1['Est. Income'], bins=bins, labels=labels, right=False)
```

```python
df1['income band'].value_counts()
df1['income band'].value_counts().plot(kind='bar')
```

```python
categorical_cols = ['BRId','GenderId','IAId','Amount of Credit Cards','Nationality','Occupation','Fee Structure','Loyalty Classification','Properties Owned','Risk Weighting','income band']
for col in categorical_cols:
    print(f"Value counts for '{col}': ")
    display(df1[col].value_counts())
```

# UNIVARIANT ANALYSIS

```python
for i, predictor in enumerate(df1[['BRId','GenderId','IAId','Amount of Credit Cards','Nationality','Occupation','Fee Structure','Loyalty Classification','Properties Owned','Risk Weighting','income band']].columns):
    plt.figure(i)
    sns.countplot(data=df1, x=predictor)
```

# BIVARIATE ANALYSIS

```python
for i, predictor in enumerate(df1[['BRId','GenderId','IAId','Amount of Credit Cards','Nationality','Occupation','Fee Structure','Loyalty Classification','Properties Owned','Risk Weighting','income band']].columns):
    plt.figure(i)
    sns.countplot(data=df1, x=predictor,hue='GenderId')
```

```python
# Histoplot of value count for different occupation
for col in categorical_cols:
    if col == 'Occupation':
        continue
    plt.figure(figsize=(15,4))
    sns.histplot(df1[col])
    plt.title("Histogram of occupation count")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
   
```

# NUMERIC ANALYSIS

```python
numeric_col = ['Est. Income','Superannuation Savings','Credit Card Balance','Bank Loans','Bank Deposits','Saving Accounts','Foreign Currency Account','Business Lending']
plt.figure(figsize=(20,23))
for i,col in enumerate(numeric_col):
    plt.subplot(4,3,i+1)
    sns.histplot(df1[col], kde=True)
    plt.title(col)
plt.show()
```

```python
# Heat map

numeric_col = ['Est. Income','Superannuation Savings','Credit Card Balance','Bank Loans','Checking Accounts','Bank Deposits','Saving Accounts','Foreign Currency Account','Business Lending']

correlation_matrix = df1[numeric_col].corr()

plt.figure(figsize=(8,8))
sns.heatmap(correlation_matrix, annot=True, cmap='crest', fmt='.2f')
plt.title('Heatmap')
plt.show()
```

# Insights of EDA

The strongest positive correlation occur amoung 'Bank deposits' with 'Checking accounts', 'Saving accounts' and 'Foreign currency account'
indicating that customers who maintain high balance in one account type often holds the substantial amount/funds accross other accounts as well.
